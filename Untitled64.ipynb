```python
import pandas as pd
import numpy as np
import os
import datetime
import sys
import itertools
from datetime import timedelta, date
import matplotlib.pyplot as plt
import seaborn as sns
import math
from statsmodels.formula.api import ols
```

```python
df = pd.read_csv('result_rog_csc_1.csv')
pd.set_option('display.float_format', lambda x: '%.2f' % x)
df.columns = [col.lower() for col in df.columns]
df.head()
```

```python
df = df.fillna(0)
df.drop(columns = ['rog_id', 'week_start_date'], axis = 1, inplace = True)
#df.set_index('week_end_date', inplace = True)
df.head()
```

```python
sns.pairplot(df, kind = 'reg')
```

```python
sns.pairplot(df, kind = 'kde')
```

```python
df_429760 = df[df.consumer_selling_cd == 429760]
df_429760.set_index('week_end_date', inplace = True)
df_429760['ln_total_item_qty'] = np.log(df_429760['total_item_qty'])
#df_429760['ln_total_amt'] = np.log(df_429760['total_amt'])
df_429760['ln_retail_price'] = np.log(df_429760['max_reg_rtl_prc'] / df_429760['max_reg_rtl_prc_fctr'])
#df_429760['ln_promo_prc'] = np.log(df_429760['max_promo_prc'] / df_429760['max_promo_prc_fctr'])
df_429760['ln_max_eff_prc'] = np.log(df_429760['max_eff_prc'] / df_429760['max_eff_prc_fctr'])
df_429760.drop(columns = ['consumer_selling_cd', 
                          'total_item_qty', 
                          'total_amt', 
                          'max_reg_rtl_prc', 
                          'max_reg_rtl_prc_fctr',
                          'max_promo_prc',
                          'max_promo_prc_fctr',
                          'max_eff_prc',
                          'max_eff_prc_fctr'
                        ], axis = 1, inplace = True)
df_429760.head()
```

```python
sns.pairplot(df_429760, diag_kind = 'kde')
```

```python
plt.figure(figsize=(10,8),dpi=100)
# Getting the Upper Triangle of the co-relation matrix
matrix = np.triu(df_429760.corr())
sns.heatmap(df_429760.corr(), cmap = "viridis", annot = True, linewidth = 0.5, mask = matrix)
```

```python
sns.pairplot(df_429760, diag_kind = 'kde')
```

```python
model_429760_robust_1 = ols("ln_total_item_qty ~ ln_max_eff_prc + ln_retail_price", data = df_429760).fit(cov_type = "HC0")
print(model_429760_robust_1.summary())
print(model_429760_robust_1.conf_int())
```

```python
model_429760_robust_2 = ols("ln_total_item_qty ~ ln_max_eff_prc", data = df_429760).fit(cov_type = "HC0")
print(model_429760_robust_2.summary())
print(model_429760_robust_2.conf_int())
```

```python
model_429760_std = ols("ln_total_item_qty ~ ln_max_eff_prc", data = df_429760).fit()
print(model_429760_std.summary())
print(model_429760_std.conf_int())
```
