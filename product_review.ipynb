As part of data cleaning process, we are removing records where we do not have any title information or the store information is contained within title.

```python
orig_df_meta = copy.deepcopy(df_meta)

df_meta = df_meta[df_meta['title'] != '']
print(len(df_meta))

df_meta = df_meta[~(df_meta['title'] == df_meta['store'])]
print(len(df_meta), len(df_meta['title'].unique()))
```

```python
df_meta['title'] = df_meta['title'].astype(str)
df_meta['title'].value_counts()
```

We see 109K records have unique title and 1435 records have duplicate titles.

```python
df_temp = df_meta['title'].value_counts().reset_index()
len(df_temp[df_temp['count'] > 1]), len(df_temp[df_temp['count'] == 1])
```
