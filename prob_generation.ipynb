```python
res=[] 
#0.03 for 90 -> this starts at 76 days
#0.03 for 95 -> this starts at 99 days
#0.05 for 99 -> this starts at 90 days

for i in np.arange(0.05,1.01,0.01):
    for n in range(1,300):
        res.append([round(i,2),n,round(pow(1-i,n),4)])
final_df = pd.DataFrame(res, columns=['item_sale_frequency', 'n_consecutive_zero_sale_days','probability'])
new_df=final_df[final_df['probability']<=0.01]
final_df = new_df.groupby('item_sale_frequency',as_index = False).first()
dis=final_df[final_df['item_sale_frequency']==0.05]
dis
```

```python
import matplotlib.pyplot as plt

# point3_df = final_df[final_df['item_sale_frequency']<=0.3]
point3_df=final_df
x = point3_df['item_sale_frequency']  # X-axis points
y = point3_df['n_consecutive_zero_sale_days'] # Y-axis points

fig, ax = plt.subplots()
# plt.plot(x, y)  # Plot the chart
ax.set_xlabel('Item Sale Frequency')
ax.set_ylabel('n-consective zero sale days for PI reset')

ax.plot(x,y)
plt.xticks(np.arange(0, 1.01, 0.1))
plt.yticks(np.arange(0, 51, 10))
# plt.figtext(.8, .8, "T=0.1")
plt.show()
```
