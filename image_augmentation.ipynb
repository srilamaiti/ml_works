```python
# Define function to show original and transformed image side by side
def show_both_images(original, transformed):
    fig, ax = plt.subplots(1,2)
    ax[0].axis('off')
    ax[1].axis('off')
    ax[0].set_title('Original')
    ax[1].set_title(f'Transformed')
    ax[0].imshow(original);
    ax[1].imshow(transformed);
    plt.show();
```

```python
# Import scikit-image
from skimage import transform

# Perform 50 degrees clockwise rotation
transformed_img = transform.rotate(sample_img,
                                   angle=-50,
                                   cval=0 # Fill with black bg
                                  )

# Compare images
show_both_images(sample_img, transformed_img);
```

```python
# Import Augmentor
import Augmentor

# Instantiate a Pipeline pointing to target input directory
p = Augmentor.Pipeline(os.getcwd() + "/input/")

# Perform horizontal flip
p.flip_left_right(probability=.3)

# We then sample (1 image) from the pipeline
num_of_samples = 1
p.sample(num_of_samples)

# Compare images
transformed_img = Image.open(image_file, 'r')
show_both_images(sample_img, transformed_img);
```

```python
# Import albumentations
import albumentations as A

# Define augmentation pipeline
transform = A.Compose([
            A.RandomBrightnessContrast(p=1.0, brightness_limit=1)
            ])

# Run pipeline on original image
transformed = transform(image=sample_img)
transformed_img = transformed["image"]

# Compare images
show_both_images(sample_img, transformed_img);
```

```python
# Import torchvision transforms
import torchvision.transforms as T

# Define Compose transformation pipeline (with random crop)
transform = T.Compose([T.RandomCrop(size=320)
                      ])

# Set input as PIL Image format (instead of numpy_array)
sample_img_PIL = Image.open(image_file)

# Run transformation pipeline on PIL image
transformed_img = transform(sample_img_PIL) 

# Compare images
show_both_images(sample_img, transformed_img);
```

```python
# Import imgaug
from imgaug import augmenters as iaa

# Define transformation (Sequential) pipeline for Gaussian noise
transform = iaa.Sequential([iaa.AdditiveGaussianNoise(scale=(20, 200))
                            ])

# Run transformation on sample image
transformed_img = transform(image=sample_img)

# Compare images
show_both_images(sample_img, transformed_img);
```

```python
image_gray = iaa.Grayscale(1.0)(image=sample_img)
show_both_images(sample_img, image_gray);
```

```python
# Import OpenCV
import cv2

# Get height and width of the image
height, width = sample_img.shape[:2]

# Set translation parameters
quarter_height, quarter_width = height/4, width/4
  
translation_matrix = np.float32([[1, 0, quarter_width], 
                                 [0, 1, quarter_height]])
  
# Use of warpAffine to perform translation with matrix
transformed_img = cv2.warpAffine(sample_img, translation_matrix, 
                                 (width, height))

# Compare images
show_both_images(sample_img, transformed_img);
```

```python
import imageio
import imgaug as ia
import imgaug.augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
```

```python
input_img = imageio.imread(image_file)
```

```python
#Horizontal Flip
hflip= iaa.Fliplr(p=1.0)
input_hf= hflip.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_hf);
```

```python
#Vertical Flip of original image
vflip= iaa.Flipud(p=1.0) 
input_vf= vflip.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_vf);
```

```python
#Vertical Flip of horizontal image
vflip= iaa.Flipud(p=1.0) 
input_vf= vflip.augment_image(input_hf)
# Compare images
show_both_images(sample_img, input_vf);
```

```python
rot1 = iaa.Affine(rotate=(-50,20))
input_rot1 = rot1.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_rot1);
```

```python
crop1 = iaa.Crop(percent=(0, 0.3)) 
input_crop1 = crop1.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_crop1);
```

```python
noise=iaa.AdditiveGaussianNoise(10,40)
input_noise=noise.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_noise);
```

```python
shear = iaa.Affine(shear=(-80,100))
input_shear=shear.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_shear);
```

```python
contrast=iaa.GammaContrast((0.5, 2.0))
contrast_sig = iaa.SigmoidContrast(gain=(5, 10), cutoff=(0.4, 0.6))
contrast_lin = iaa.LinearContrast((0.6, 0.4))

input_contrast = contrast.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_contrast);

sigmoid_contrast = contrast_sig.augment_image(input_img)
# Compare images
show_both_images(sample_img, sigmoid_contrast);

linear_contrast = contrast_lin.augment_image(input_img)
# Compare images
show_both_images(sample_img, linear_contrast);
```

```python
elastic = iaa.ElasticTransformation(alpha=60.0, sigma=4.0)
polar = iaa.WithPolarWarping(iaa.CropAndPad(percent=(-0.2, 0.7)))
jigsaw = iaa.Jigsaw(nb_rows=20, nb_cols=15, max_steps=(3, 7))

input_elastic = elastic.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_elastic);

input_polar = polar.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_polar);

input_jigsaw = jigsaw.augment_image(input_img)
# Compare images
show_both_images(sample_img, input_jigsaw);
```
