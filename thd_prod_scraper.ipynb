```python
# set up the request parameters
params = {
'api_key': '46B885F03D8C44F59E438123EA4DC0D2',
  'search_term': 'drill',
  'type': 'search'
}

# make the http GET request to BigBox API
api_result = requests.get('https://api.bigboxapi.com/request', params)

# print the JSON response from BigBox API
print(json.dumps(api_result.json()))
```

```python
prod_list_drill = api_result.json()["search_results"]
```

```python
prod_attr_list = []
for prod in prod_list_drill:
    try:
        model_name = prod['product']['model_number']
    except:
        model_name = None

    try:
        model_desc = "'" + prod['product']['title'] + "'"
    except:
        model_desc = None

    try:
        model_brand = "'" + prod['product']['brand'] + "'"
    except:
        model_brand = None

    try:
        model_type = "'" + prod['product']['features'][0]['value'] + "'"
    except:
        model_type = None

    try:
        model_motor_type = "'" + prod['product']['features'][1]['value'] + "'"
    except:
        model_motor_type = None

    try:
        model_chuck_size = "'" + prod['product']['features'][2]['value'] + "'"
    except:
        model_chuck_size = None

    try:
        model_motor_voltage = "'" + prod['product']['features'][3]['value'] + "'"
    except:
        model_motor_voltage = None

    try:
        model_sale_price = "'" + str(prod['offers']['primary']['price']) + " USD'"
    except:
        model_sale_price = None

    try:
        model_regular_price = "'" + str(prod['offers']['primary']['regular_price']) + " USD'"
    except:
        model_regular_price = None

    try:
        model_rating = "'" + str(prod['product']['rating']) + "'"
    except:
        model_rating = None

    try:
        model_total_rating_count = "'" + str(prod['product']['ratings_total']) + "'"
    except:
        model_total_rating_count = None

    prod_attr_list.append((model_name, model_desc, model_brand, model_type, model_motor_type, model_chuck_size, model_motor_voltage, model_sale_price, model_regular_price, model_rating, model_total_rating_count))
```

```python
df_drill = pd.DataFrame(prod_attr_list)
df_drill.columns = ['model_name', 'model_desc', 'model_brand', 'model_type', 'model_motor_type', 'model_chuck_size', 'model_motor_voltage', 'model_sale_price', 'model_regular_price', 'model_rating', 'model_total_rating_count']
df_drill = df_drill.sort_values(by = ['model_rating', 'model_total_rating_count'], ascending = [False, False])
df_drill
```
