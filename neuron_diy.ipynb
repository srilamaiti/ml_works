```python
from typing import Callable
```

```python
Vector = [float]
ActivationFunction = Callable[[float], float]
```

image.png

```python
def neuron(inputs: Vector, weights: Vector, activation_func: ActivationFunction) -> float:
	return activation_func(
		sum(
			z[0] * z[1] for z in zip([1.0] + inputs, weights)
		)
	)

def step(x: float) -> float:
	return 1 if x > 0 else 0

# weights are richness and smokieness
weights = [-0.25, 1, -0.45]

glenmorangie = [-0.21, 0.18]
talisker = [0.6, -0.31]

print("Glenmorangie:", neuron(glenmorangie, weights, step))
print("Talisker:", neuron(talisker, weights, step))
```

```python
NAND = [1.5, -1, -1]
OR = [-0.5, 1, 1]
AND = [-1, 1, 1]

def xor_net(inputs: Vector) -> float:
	return neuron(
		inputs=[
			neuron(inputs, NAND, step),
			neuron(inputs, OR, step)
		],
		weights=AND,
		activation_func=step
	)

inputs = [
	[0, 0],
	[1, 0],
	[0, 1],
	[1, 1]
]

for i in inputs:
	print(i, xor_net(i))
```
